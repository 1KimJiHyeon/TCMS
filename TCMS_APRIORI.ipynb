```
df = pd.DataFrame(dataset)
df.head(10)
```

```
te = TransactionEncoder() # TransactionEncoder는 파이썬 리스트의 트랜잭션 데이터에 대한 인코더 클래스를 정의
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```
df
```

```
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True) #지지도 0.5이상
frequent_itemsets.sort_values(by="support", ascending=False)
```

```
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8) # 신뢰도 0.8
print(rules)
rules.to_csv("275_result.csv", mode='w')
```

```
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.05)
print(rules)
```
